```python
# import necessary packages
import numpy as np
import pandas as pd
```

```python
# import and display payment datasets
pay2020 = pd.read_csv('payment2020.csv', nrows=6000)
pay2021 = pd.read_csv('payment2021.csv', nrows=6000)
```

```python
pay2020.info()
```

```python
pay2021.info()
```

```python
# concatenate both dataframes into one and display
pay = pd.concat([pay2020, pay2021], axis=0)

pay
```

```python
pay.describe(include='all')
```

```python
# remove duplicate data
pay.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=False)

pay.duplicated().sum()
```

```python
# check for missing data
pay.isna().sum()
```

```python
# drop columns with too many missing values
pay_list = list(pay.columns.values)

for col in pay_list:
    if(pay[col].isna().sum() > 500):
        pay.drop(col, axis=1, inplace=True)
        
pay.info()
```

```python
# drop rows with missing values
pay.dropna(axis=0, inplace=True)

pay.info()
```

```python
# drop unnecesary columns
unneeded = ['change_type', 
            'covered_recipient_type',
            'covered_recipient_npi',
            'submitting_applicable_manufacturer_or_applicable_gpo_name',
            'third_party_payment_recipient_indicator',
            'delay_in_publication_indicator',
            'applicable_manufacturer_or_applicable_gpo_making_payment_id',
            'dispute_status_for_publication',
            'related_product_indicator',
            'program_year',
            'payment_publication_date']

for col in unneeded:
    pay.drop(col, axis=1, inplace=True)
        
pay.info()
```

```python
# rename certain columns
pay.rename({'covered_recipient_profile_id':'physician_id',
            'covered_recipient_first_name':'physician_first_name',
            'covered_recipient_last_name':'physician_last_name',
            'recipient_primary_business_street_address_line1':'physician_street',
            'recipient_city':'physician_city',
            'recipient_state':'physician_state',
            'recipient_zip_code':'physician_zip_code',
            'recipient_country':'physician_country',
            'covered_recipient_primary_type_1':'physician_primary_type',
            'covered_recipient_specialty_1':'physician_specialty',
            'covered_recipient_license_state_code1':'physician_license_state',
            'record_id':'payment_id',
            'applicable_manufacturer_or_applicable_gpo_making_payment_name':'payment_name',
            'applicable_manufacturer_or_applicable_gpo_making_payment_state':'payment_state',
            'applicable_manufacturer_or_applicable_gpo_making_payment_country':'payment_country',
            'total_amount_of_payment_usdollars':'payment_amount',
            'date_of_payment':'payment_date',
            'number_of_payments_included_in_total_amount':'payment_number',
            'form_of_payment_or_transfer_of_value':'payment_form',
            'nature_of_payment_or_transfer_of_value':'payment_nature'}, axis=1, inplace=True)

pay.info()
```

```python
# specify types for each field
pay_int = ['physician_id','payment_id','payment_number']
pay_char = ['physician_first_name','physician_last_name','physician_street',
            'physician_city','physician_state','physician_country',
            'physician_primary_type','physician_specialty',
            'physician_license_state','payment_name',
            'payment_state','payment_country','payment_form','payment_nature']
pay_float = ['payment_amount']
```

```python
# convert columns to appropriate types
for col in pay_int:
    pay[col] = pay[col].astype('int')
    
for col in pay_char:
    pay[col] = pay[col].astype('string')
    
for col in pay_float:
    pay[col] = pay[col].astype('float')
    
pay['payment_date'] = pd.to_datetime(pay['payment_date'], errors='coerce', format='%m/%d/%Y')

pay.dtypes
```

```python
# shorten zip to first 5 digits
pay['physician_zip_code'] = pay['physician_zip_code'].astype('string')

pay['physician_zip_code'] = pay['physician_zip_code'].str[:5]

pay['physician_zip_code'] = pay['physician_zip_code'].astype('int')

pay
```

```python
# save to local directory as CSV
pay.to_csv('payment_data.csv')
```

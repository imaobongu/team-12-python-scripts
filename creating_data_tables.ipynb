```python
# import necessary packages
import sqlite3
import numpy as np
import pandas as pd
```

```python
# import and display datasets
pay = pd.read_csv('payment_data.csv')
rate = pd.read_csv('rating_data.csv')
```

```python
pay.head()
```

```python
rate.head()
```

```python
physician = pay[['physician_id','physician_first_name','physician_last_name',
                 'physician_primary_type','physician_specialty','physician_license_state']]

location = pay[['physician_id','physician_street','physician_city','physician_state',
                'physician_country','physician_zip_code']]

payment = pay[['payment_id','payment_form','payment_nature']]

rating = rate[['rating_id','rating_source']]
```

```python
location
```

```python
PaymentPhysician = pay[['payment_id','payment_name','payment_state',
                        'payment_country','payment_amount','payment_date',
                        'payment_number']]

RatingPhysician = rate[['rating_id','rating_last_name','rating_first_name',
                        'quality_score','interoperability_score',
                        'improvement_score','mips_score']]
```

```python
PaymentPhysician
```

```python
conn = sqlite3.connect('team_12_database.db')
c = conn.cursor()

physician.to_sql('physician_table', conn, if_exists='replace', index=False)

location.to_sql('location_table', conn, if_exists='replace', index=False)

payment.to_sql('payment_table', conn, if_exists='replace', index=False)

rating.to_sql('rating_table', conn, if_exists='replace', index=False)

PaymentPhysician.to_sql('payment_physician_table', conn, if_exists='replace', index=False)

RatingPhysician.to_sql('rating_physician_table', conn, if_exists='replace', index=False)

conn.commit()
```

```python
pd.read_sql('select * from payment_table', conn)
```

```python
# import necessary packages
import numpy as np
import pandas as pd
```

```python
# import and display rating datasets
rate = pd.read_csv('ec_score_file.csv', nrows=20000)

rate.head()
```

```python
rate.info()
```

```python
rate.describe(include='all')
```

```python
# remove duplicate data
rate.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=False)

rate.duplicated().sum()
```

```python
# check for missing data
rate.isna().sum()
```

```python
# drop columns with too many missing values
col_list = list(rate.columns.values)

for col in col_list:
    if(rate[col].isna().sum() > 10000):
        rate.drop(col, axis=1, inplace=True)
        
rate.info()
```

```python
# drop rows with missing values
rate.dropna(axis=0, inplace=True)

rate.info()
```

```python
# drop unnecessary columns
unneeded = [' Org_PAC_ID',
            ' final_MIPS_score_without_CPB']

for col in unneeded:
    rate.drop(col, axis=1, inplace=True)

rate.info()
```

```python
# rename certain columns
rate.rename({'NPI':'rating_id',
             ' lst_nm':'rating_last_name',
             ' frst_nm':'rating_first_name',
             ' source':'rating_source',
             ' Quality_category_score':'quality_score',
             ' PI_category_score':'interoperability_score',
             ' IA_category_score':'improvement_score',
             ' final_MIPS_score':'mips_score'}, axis=1, inplace=True)

rate.head()
```

```python
# specify types for each field
rate_int = ['rating_id']

rate_char = ['rating_last_name','rating_first_name','rating_source']

rate_float = ['quality_score','interoperability_score','improvement_score','mips_score']
```

```python
# convert columns to appropriate types
for col in rate_int:
    rate[col] = rate[col].astype('int')
    
for col in rate_char:
    rate[col] = rate[col].astype('string')
    
for col in rate_float:
    rate[col] = rate[col].astype('float')

rate.dtypes
```

```python
# diplay
rate.head()
```

```python
# save to local directory as CSV
rate.to_csv('rating_data.csv')
```
